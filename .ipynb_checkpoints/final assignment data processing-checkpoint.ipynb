<h2>2&3. Converting all flat structures into nested structures</h2>

```python
organised_list = [] # declared to append stuctured nested data
noDependats_persons = [] # declared to append persons with no dependants
```

<p>Creating for loop for using enumerate function to access index of iterating list. First I have declared person dictionary for assign data stuctured data iterating on acw_user_list. A vehical dictionay to assign vehical related data by retriving through index by passing column name from header list, similarly with credit card and address</p>

<p>For Task 1 part: I use if else condition, if list have dependant assign it to dependants else if null assign "no dependants"</p>

<p>At the end append person the stuctured_list</p>

```python
for index, row in enumerate(acw_user_list):
    # declare person dictionary to assigned stuctered data from iterated list
    person = dict()
    
    # assigning vehical related data to vehical dict from list of rows
    vehicle = {
        'make':row[header.index('Vehicle Make')],
        'year':row[header.index('Vehicle Year')],
        'type':row[header.index('Vehicle Type')]
    }
    
    # assigning credit card related data credit_card dict from list of rows
    credit_card = {
       'start date':row[header.index('Credit Card Start Date')],
        'expiry':row[header.index('Credit Card Expiry Date')],
        'number':int(row[header.index('Credit Card Number')]),
        'ccv':int(row[header.index('Credit Card CVV')]),
        'iben':row[header.index('Bank IBAN')]
        }
    
    # assigning address related data address dict from list of rows
    address = {
        'street':row[header.index('Address Street')],
        'city':row[header.index('Address City')],
        'postcode':row[header.index('Address Postcode')]
        }
    
    # Assigning all the data of person to person dictionery delared before in a stuctured format
    person['first name'] = row[header.index('First Name')]
    person['last name'] = row[header.index('Last Name')]
    person['age'] = int(row[header.index('Age (Years)')])
    person['sex'] = row[header.index('Sex')]
    person['retired'] = row[header.index('Retired')]
    person['marital_status'] = row[header.index('Marital Status')]
    
    # Task 1 Part 3: Some person have no dependants
    # so I put condition if a person has no dependant simply assign "no dependants" string to it in the else block
    if row[header.index('Dependants')]:
        person['dependants'] = int(row[header.index('Dependants')])
    else:
        person['dependants'] = 'No dependants'
        
    # append the index to the list of person_with_noDependants to print it later
        noDependats_persons.append(index)
    
    person['salary'] = int(row[header.index('Yearly Salary (GBP)')])
    person['pension'] = int(row[header.index('Yearly Pension (GBP)')])
    person['company'] = row[header.index('Employer Company')]
    person['commute_distance'] = float(row[header.index('Distance Commuted to Work (miles)')])
    
    # assigning vehical dict to person['vehical']
    person['Vehicale'] = vehicle
    person['Credit Card'] = credit_card
    person['Address'] = address
    
    # Append person dictionary with assigned data from the above code
    organised_list.append(person)
```

<h2>6. Writing Function check whether the row should fledged if card start and end date difference is greater than 10</h2>

```python
# writing fladged_person function with row parameter.
# The function is reading person from csv file and get the row passed in the function
# After calculating total difference if it's greater than 10 function with return the row
def fladge_person(row):
    # Reading acw_user_data.csv
    with open('acw_user_data.csv', mode='r') as user_file:
        csv_reader = csv.reader(user_file)

        #assigning head row to header variable
        header = next(csv_reader)

        #assigining data to csv_reader data to data_list with skipped header row
        data_list = list(csv_reader)

        #assigning start and end card date item from rows to start_date and end_date
        start_date = (data_list[row][header.index('Credit Card Start Date')])
        end_date = (data_list[row][header.index('Credit Card Expiry Date')])

        #using split function to split date into month and year into a list
        start_date = start_date.split('/')
        end_date = end_date.split('/')

        #assigning month and year from start_date list index 0 and 1 to start_month and start_year and similarly for end_date
        start_month, start_year = int(start_date[0]), int(start_date[1])
        end_month, end_year = int(end_date[0]), int(end_date[1])

        #Calculating year_difference
        #if start month is greater than end month subtract 1year from end year as a carry
        #add 1 year(12 months) carry from end year to end month
        if start_month > end_month:
            end_year = end_year -1
            end_month = end_month + 12


        month_difference = end_month - start_month #subracting end_month from start_month and assign to month_difference
        year_difference = end_year - start_year #subracting end_year from start_year and assign to year_difference

        # Hence calculating total months by multipling 12 to year_difference and add to month_difference
        # Dividing total months by 12 to calculate total difference
        total_difference = (month_difference + (year_difference*12))/12

        #if the total difference in greater then 10 return the index and print the statement
        if total_difference > 10:
            print("Write the Current person to remove_ccard.json because it has Credit Card year difference greater than 10 !!\n")
            return row
        else:
            print("No need to write to remove_ccard.json")
```

```python
remove_ccard = [] # initialize empty row

#taking Input a person row number from user
person_row = int(input("Enter Row Number : "))

#passing person_row and storing the returning value
person_fladged = fladge_person(person_row)

# if the return value from function flage_person is not none append it to remove_ccard list
if person_row:
    remove_ccard.append(organised_list[person_fladged])
    #writing the fladged person the remove_ccard.json
    with open('remove_ccard.json', mode='w') as wj:
        json.dump(remove_ccard, wj)
```

<h2>7. Creating new data attribute Salary-Commute</h2>

```python
# reading processed.json file and assign to processed_list loading from json
with open('processed.json', mode='r') as read_file:
    processed_list = json.load(read_file)
```

```python
# Iterat over processed_list using for loop
# if commute_distance is greater than 1 divide salary by commute_distance assign to new attribut Salary-Commute
#else assign orignal salary to the new attribut Salary-Commute
for row in processed_list:
    if row['commute_distance'] > 1:
        row['Salary-Commute'] = row['salary']/row['commute_distance']
    else:
        row['Salary-Commute'] = row['salary']
```

```python
# Sort the processed_list using sorted function by passing processed_list and Salary-Commute attribute value as lambda function
sorted_processed_list = sorted(processed_list, key=lambda x: x['Salary-Commute'])
```

```python
# Writing new file as commute.json and dump the sorted processed list
with open('commute.json', mode='w') as w_commute:
    json.dump(sorted_processed_list , w_commute)
```
